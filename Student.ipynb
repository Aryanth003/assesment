```python
df[df['student_id']==14061]
```

```python
df.drop_duplicates(inplace=True)
```

### Data Transformation

```python
grade_map={
    'a': 1,
    'b':2,
    'c':3,
    'd':4,
    'e':5,
    'f':6
}
df['final_grade']=df['final_grade'].map(grade_map)
```

```python
df['extra_activities']=df['extra_activities'].map({'yes':1,'no':0})
```

```python
df['internet_access']=df['internet_access'].map({'yes':1,'no':0})
```

```python
df['gender'].value_counts()
```

### distribution

```python
fig,ax=plt.subplots(1, 3, figsize=(20, 6))
for j,i in enumerate(['age','attendance_percentage','study_hours']):
  sns.histplot(df[i],ax=ax[j])
  ax[j].set_title(f"distribution of {i}")
plt.tight_layout()
plt.show()
```
