```python
f = open("demofile.txt")
f = open("demofile.txt", "rt")
content = f.read()
print(content)
f.close()



# Alternatively, using 'with' to handle file closing automatically
with open("demofile.txt", "rt") as f:
    content = f.read()
    print(content)

# Open a file in read text mode and print its content
```
